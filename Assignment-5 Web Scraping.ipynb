```python
# Importing the Libraries
import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time
start_page=0
end_page=8
#Loading Driver
driver=webdriver.Edge(r"C:\Users\PANKAJ\Downloads\msedgedriver.exe")
# Giving the name of the website which we have to scrap
url="https://in.indeed.com/jobs?q=data+scientist&l=delhi"
# Getting the URL
driver.get(url)
# we have to scrap the below attached details
Job_Name=[]
Job_Location=[]
Company_Name=[]
Salary=[]
Job_Desc=[]
Post_Days=[]
Exp_Years=[]
# now lets scrap the data from the server            
name=driver.find_elements_by_xpath("//h2[@class='title']")
for n in name:
    j_name=n.text
    Job_Name.append(j_name)
    JN=pd.DataFrame(Job_Name)
    JN=JN.rename({0: 'Job Name'},axis=1)
location=driver.find_elements_by_xpath("//span[@class='location accessible-contrast-color-location']")
for ln in location:
    ln_name=ln.text
    Job_Location.append(ln_name)
    LN=pd.DataFrame(Job_Location)
    LN=LN.rename({0: 'Job Location'},axis=1)
company=driver.find_elements_by_xpath("//div[@class='sjcl']/div[1]")
for co in company:
    co_name=co.text
    Company_Name.append(co_name)
    CN=pd.DataFrame(Company_Name)
    CN=CN.rename({0: 'Company Name'},axis=1)
salary=driver.find_elements_by_xpath("//span[@class='salary no-wrap']")
for s in salary:
    s_amt=s.text
    Salary.append(s_amt)
    S=pd.DataFrame(Salary)
    S=S.rename({0: 'Salary P/M/Y'},axis=1)
desc=driver.find_elements_by_xpath("//div[@class='summary']")
for d in desc:
    d_short=d.text
    Job_Desc.append(d_short)
    JD=pd.DataFrame(Job_Desc)
    JD=JD.rename({0: 'Job Description'},axis=1)
postday=driver.find_elements_by_xpath("//span[@class='date ']")
for p in postday:
    post=p.text
    Post_Days.append(post)
    PD=pd.DataFrame(Post_Days)
    PD=PD.rename({0: 'Post Days'},axis=1)
    
# now lets create the data table and adding the columns into the table
indeed=pd.concat([JN,LN,CN,S,JD,PD],axis=1)
indeed
```

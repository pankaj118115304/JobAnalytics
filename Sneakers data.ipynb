```python
# Importing Libraries 

import numpy as np 
import  pandas as pd
from selenium import webdriver
from time import  sleep
import time
import string
from random import randint
import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Edge(r"C:\Users\PANKAJ\Downloads\msedgedriver.exe")
# name of the website that we have to  scrap
url = "https://www.flipkart.com/search?q=sneakers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
# Getting Url
driver.get(url)
```

```python
# we are finding the below attached details from the web site
Brand = []
Product_Description = []
Price = []
Discount = []
```

```python
# lets finidng the name of the brands 
Task4 = driver.find_elements_by_xpath("//div[@class='_2B_pmu']")
Task4
```

```python
# lets scrap the brand names from the tag with the help of for loop 
for i in Task4:
    qz = i.text
    Brand.append(qz)
```

```python
# lets create the dataframe of that brands
Brand1 = pd.DataFrame(Brand)
Brand1 = Brand1.rename(columns = { 0 : 'Brand'})
Brand1
```

```python
# lets finding the product description of the brands
Task5 = driver.find_elements_by_xpath("//div[@class='_2LFGJH']/a[1]")
Task5
```

```python
# lets scrap the product desc from the tag with the help of for loop
for iz in Task5:
    qzp = iz.text
    Product_Description.append(qzp)
```

```python
# lets create the dataframe of product description table
Product_Description1 = pd.DataFrame(Product_Description)
Product_Description1 = Product_Description1.rename({ 0: 'Product Description'} , axis=1 )
Product_Description1
```

```python
Task6 = driver.find_elements_by_xpath("//div[@class='_1uv9Cb']/div[1]")
Task6[0:5]
```

```python
for apv in Task6:
    apx =apv.text
    Price.append(apx)
```

```python
Price1 = pd.DataFrame(Price)
Price1 = Price1.rename(columns = { 0 : 'Price'})
Price1
```

```python
Price2 = Price1.iloc[0:80]
Price2
```

```python
# lets finding the discount offered of the brands
Task7 = driver.find_elements_by_xpath("//div[@class='VGWI6T']/span")
Task7
```

```python
# lets scrap the discounts offered on the brands with the help of for loop
for i in Task7:
    dof = i.text
    Discount.append(dof)
```

```python
# lets create the discount table
Discount1 = pd.DataFrame(Discount)
Discount1 =Discount1.rename(columns = { 0 : 'Discount'})
Discount1
```

```python
# finding the length of each attribute
print(len(Brand1),len(Product_Description1),len(Price2),len(Discount1))
```

```python
# lets add the whole data in a table
Sneakers1 =pd.concat([Brand1,Product_Description1,Price2,Discount1] , axis=1)
Sneakers1
```

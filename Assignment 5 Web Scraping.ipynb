```python
# importing Required library 

from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
import time
from datetime import date

driver=webdriver.Edge(r"C:\Users\PANKAJ\Downloads\msedgedriver.exe")
driver.get("https://in.indeed.com/jobs?q=Data+Scientist&l=Delhi")
# finding the details which are scraped from the web server
jobtitle = []
company = []
location = []
daysposted = []
salary = []
description = []
crawl_timestamp = []
crawl_date = []
```

```python
jobs = driver.find_elements_by_class_name('result')
# lets scrap the data from the server with the help of for loop
for job in jobs:
    
    res= job.get_attribute('innerHTML')
    jb = BeautifulSoup(res,'html.parser')

    try:
        title = jb.find("a" ,class_="jobtitle").text.replace('\n',' ')
        jobtitle.append(title)
    except:
        jobtitle.append('  ')
 
    try:
        com= jb.find("span" ,class_="company").text.replace('\n',' ')
        company.append(com)
    except:
        company.append(' ')

    try:
        lo = jb.find("span" ,class_="location accessible-contrast-color-location").text.replace('\n',' ')
        location.append(lo)
    except:
        location.append(' ')      
      
    try:
        sal = jb.find("span" ,class_="salaryText").text.replace('\n',' ')
        salary.append(sal)
    except:
        salary.append('Not Mentioned')
       
    try:
        day = jb.find("span" ,class_="date").text.replace('\n',' ')
        daysposted.append(day)
    except:
        daysposted.append(' ')
     
    try:
        jo = jb.find("div" ,class_="summary").text.replace('\n',' ')
        description.append(jo)
    except:
        description.append('----')
# lets scrap the date time form the server with the help of loop            
    time.sleep(2)
            
    task1 = time.gmtime()
    crawl_timestamp.append(time.strftime("%Y-%m-%d %H:%M:%S", task1))

    today = date.today()    
    task2 = today.strftime("%d %B %Y")
    crawl_date.append(task2)       
```

```python
# lets finding the leangth of the scraped attributes so that we can create the dataframe of the result 
print(len(jobtitle),len(company),len(location),len(daysposted),len(salary),len(description),len(crawl_timestamp),len(crawl_date))
```

```python
# lets create the indeed data table
indeed =pd.DataFrame({ 'Job Name' : jobtitle ,'Job Location' : location,'Company Name' : company ,' Salary' :salary ,'Post Days' : daysposted ,'Time': crawl_timestamp , ' Date' : crawl_date ,'Job Description' : description  })
indeed
```

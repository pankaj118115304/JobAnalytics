# Removing the Outliers and Skewness

```python
whr.skew()
```

```python
from scipy.stats import zscore
z=np.abs(zscore(whr))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
WHR=whr[(z<3).all(axis=1)]
WHR
```

```python
print(whr.shape,WHR.shape)
```

```python
for col in WHR.columns:
    if WHR.skew().loc[col]>0.55:
        WHR[col]=np.log1p(WHR[col])
```

```python
WHR.skew()
```

```python
WHR.info()
```

# Spliting the data into input and target

```python
x=WHR.drop('Happiness Score',axis=1)
x.head()
```

```python
y=pd.DataFrame(WHR['Happiness Score'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg=LinearRegression()
lg.fit(x_train,y_train)
```

```python
lg.coef_
```

```python
lg.intercept_
```

```python
lg.score(x_train,y_train)
```

```python
# to find best fit random state
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(42,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r_state)
    lrg=linear_model.LinearRegression()
    lrg.fit(x_train,y_train)
    y_pred=lrg.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print('r2_score coorsponding to random state is ',r_state,'is:',r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
    
print()
print()
print('Max r2 is cooresponding to ',final_r_state,'is:',max_r_score)
```

```python
pred=lg.predict(x_test)
print('Predicted Result Price is ',pred)
print('Actual Price is ',y_test)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=68,test_size=.20)
lrg=linear_model.LinearRegression()
lrg.fit(x_train,y_train)
lrg_pred=lrg.predict(x_test)
print('r2_score is ',r2_score(y_test,lrg_pred))
print('RSME is ',np.sqrt(mean_squared_error(y_test,lrg_pred)))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring='r2')
```

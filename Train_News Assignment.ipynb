```python
# filling the null values to add the news in total_massege columns
tn=tn.fillna(' ')
tn['total_message']=tn['headline']+' '+tn['written_by']+tn['news']
```

```python
# deleting the headline , written_by and news columns in the data because we are adding the whole news in one column i.e. total_message
tn.drop(columns=['headline','written_by','news','id'],axis=1,inplace=True)
```

```python
# finding the fake news and true news data
tn.label.value_counts()
```

```python
tn['Length']=tn.total_message.str.len()
tn
```

```python
# coverting the whole news message into smaller text
tn['total_message']=tn['total_message'].str.lower()
```

```python
# replace money symbols with 'moneysymb'
tn['total_message']=tn['total_message'].str.replace(r'$','dollars')
# Replace URLs with webaddress
tn['total_message']=tn['total_message'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$','webaddress')
# replace numbers as well 
tn['total_message']=tn['total_message'].str.replace(r'\d+(\.\d+)?','number')
# replace phone numbers 
tn['total_message']=tn['total_message'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$','phonenumber')
```

```python
# removing the punctuation
tn['total_message']=tn['total_message'].str.replace(r'[^\w\d\s]', ' ')
# replace white spcae with single space
tn['total_message']=tn['total_message'].str.replace(r'\s+',' ')
# remove leading or trailing white space 
tn['total_message']=tn['total_message'].str.replace(r'^\s+|\s+?$','')
```

```python
import string
import nltk
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english')+ ['u','ur','4','2','im','dont','doin','ure','ok'])
tn['total_message']=tn['total_message'].apply(lambda x:' '.join(w for w in x.split() if w not in stop_words))
```

```python
tn['NLength']=tn.total_message.str.len()
tn
```

```python
# finding the length of orignal data and cleaned data
print('Original Length',tn.Length.sum())
print('Clean Data Length',tn.NLength.sum())
```

# Predicting the model with various modeling techniques

```python
# convert all text into vectors using TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
message=tf.fit_transform(tn['total_message'])
x=message
y=tn['label']
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score , GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier 
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lgr=LogisticRegression()
#sv=SVC()
dtc=DecisionTreeClassifier()
kn=KNeighborsClassifier()
rfc=RandomForestClassifier()
abc=AdaBoostClassifier()
gbc=GradientBoostingClassifier()
models=[]
models.append(('LogisticRegression',lgr))
#models.append(('SVC',sv))
models.append(('DecisionTreeClassifier',dtc))
models.append(('RandomForestClassifier',rfc))
models.append(('AdaBoostClassifier',abc))
models.append(('GradientBoostingClassifier',gbc))
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
for i,model in models:
    print('******************',i,'*******************','\n')
    Model.append(i)
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    AS=accuracy_score(y_test,pre)
    print('Accuracy Score is ',AS,'\n')
    score.append(AS*100)
    CM=confusion_matrix(y_test,pre)
    print('Confusion Matrix is \n',CM,'\n')
    print('Classification Report is',classification_report(y_test,pre),'\n')
    cs=cross_val_score(model,x,y,cv=5,scoring='accuracy').mean()
    print('Cross Validation Score is ',cs,'\n')
    cvs.append(cs*100)
    fpr,tpr,thresholds=roc_curve(y_test,pre)
    roc_auc=auc(fpr,tpr)
    print('Roc Auc Score is',roc_auc,'\n')
    rocscore.append(roc_auc*100)
```

```python
result=pd.DataFrame({'Model':Model,'Accuracy Score':score,'CVS':cvs,'ROC AUC':rocscore})
result
```

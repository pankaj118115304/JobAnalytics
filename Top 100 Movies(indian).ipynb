```python
from bs4 import BeautifulSoup
import requests
ind=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
soup=BeautifulSoup(ind.content,'html.parser')
```

```python
# we are scraping the name , ratinf and  year of release
Name=[]
Rating=[]
YearofRelease=[]
```

```python
soup.title.text
```

```python
movies=soup.find('tbody',class_="lister-list").findAll("tr")
movies
```

```python
first_movie=movies[0]
```

```python
first_movie.find('td',class_="titleColumn").a.text
```

```python
first_movie.find('span',class_="secondaryInfo").text[1:5]
```

```python
first_movie.find('td',class_="ratingColumn imdbRating").strong.text
```

```python
for i in movies:
    Name.append(i.find('td',class_="titleColumn").a.text)
    YearofRelease.append(i.find('span',class_="secondaryInfo").text[1:5])
    Rating.append(i.find('td',class_="ratingColumn imdbRating").strong.text)
```

```python
data=list(zip(Name,YearofRelease,Rating))
```

```python
import pandas as pd
ind=pd.DataFrame(data,columns=['Name','Release_Year','Rating'])
ind[0:100]
```

```python
from bs4 import BeautifulSoup
import requests

# now i will sending the request to webpage server to get the source code of the page which is to to scraped
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
# see the source code of the webpage
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
titles=soup.find_all(['h1','h2','h3','h4','h5','h6'])
print('List of all header tahes are : \n ',titles, sep='\n')
```

# Scrap the Data from book page .com 

```python
from bs4 import BeautifulSoup
bp=requests.get('https://bookpage.com/')
soup=BeautifulSoup(bp.content,'html.parser')
```

```python
book_name=[]
author_name=[]
genre_name=[]
```

```python
titles=soup.find_all('h4',class_="italic")
titles[0:5]
```

```python
for i in titles:
    book_name.append(i.get_text())
book_name[0:5]    
```

```python
author=soup.find_all('p',class_="sans bold")
author[0:5]
```

```python
for i in author:
    author_name.append(i.get_text())
author_name[0:5]    
```

```python
genre=soup.find_all('p',class_="genre-links hidden-phone")
genre[0:5]
```

```python
for i in genre:
    genre_name.append(i.get_text())
genre_name[0:5]
```

```python
print(len(book_name),len(author_name),len(genre_name))
```

```python
import pandas as pd
book=pd.DataFrame({})
book['Title']=book_name
book['Author']=author_name
book['Genre']=genre_name
book
```

# Top 10 ODI team in men's Cricket , with their matches , rating and points

```python
from bs4 import BeautifulSoup
odi=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
soup=BeautifulSoup(odi.content,'html.parser')
```

```python
# we are finding the top 10 teams ,matches,points and ratings
Teams=[]
Matches=[]
Points=[]
Rating=[]
```

```python
team=soup.find_all('span',class_="u-hide-phablet")
team[0:10]
```

```python
for i in team:
    Teams.append(i.get_text())
Teams[0:10]    
```

```python
match=soup.find_all('td',class_="rankings-block__banner--matches")
match[0:10]
```

```python
for i in match:
    Matches.append(i.get_text())
Matches[0:10]
```

```python
rating=soup.find_all('td',class_="rankings-block__banner--rating u-text-right")
rating[0:10]
```

```python
for i in rating:
    Rating.append(i.get_text())
Rating[0:10]
```

```python
point=soup.find_all('td',class_="rankings-block__banner--points")
point[0:10]
```

```python
for i in point:
    Points.append(i.get_text())
Points[0:10]
```

# Write a program to extract the top 100 movies name and IMDB rating and Year of Release

```python
from bs4 import BeautifulSoup
imdb=requests.get("https://www.imdb.com/search/title/?groups=top_250&sort=user_rating")
soup=BeautifulSoup(imdb.content,'html.parser')
```

```python
Name=[]
IMDB_Rating=[]
Release_Year=[]
```

```python
imdb.content
```

```python
soup.title.text
```

```python
name=soup.find_all('div',class_="lister-item mode-advanced")
name
```

```python
first_movie=name[0]
```

```python
first_movie.h3.a.text
```

```python
first_movie.find('span',class_="lister-item-year text-muted unbold").text[1:5]
```

```python
first_movie.find('div',class_="inline-block ratings-imdb-rating")['data-value']
```

```python
for i in name:
    Name.append(i.h3.a.text)
    Release_Year.append(i.find('span',class_="lister-item-year text-muted unbold").text[1:5])
    IMDB_Rating.append(i.find('div',class_="inline-block ratings-imdb-rating")['data-value'])
```

```python
data=list(zip(Name,Release_Year,IMDB_Rating))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=['Name','Year','Rating'])
```

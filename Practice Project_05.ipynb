# Removing the skewness and outliers from the data

```python
hd.skew()
```

```python
from scipy.stats import zscore
z=np.abs(zscore(hd))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
hdn=hd[(z<3).all(axis=1)]
hdn
```

```python
print(hd.shape , hdn.shape)
```

```python
for col in hdn.columns:
    if hdn.skew().loc[col]>0.55:
        hdn[col]=np.log1p(hdn[col])
```

```python
hdn.skew()
```

# Seprating Input and Output Variables from the data

```python
x=hdn.iloc[:,0:-1]
x.head()
```

```python
y=hdn.iloc[:,-1]
y.head()
```

```python
# To training the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import cross_val_score
max_r_score=0
for r_state in range(42,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r_state)
    lr.fit(x_train,y_train)
    pre=lr.predict(x_test)
    r2_scr=r2_score(y_test,pre)
    print('R2 score is ',r_state,'is',r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
        
print()
print('Max R2 Score is ',final_r_state,'is:',max_r_score)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=96)
lr.fit(x_train,y_train)
pre=lr.predict(x_test)
print('r2 score is',r2_score(y_test,pre))
print('RMSE is ',np.sqrt(mean_squared_error(y_test,pre)))
print('Cross Validation Score is ',cross_val_score(LinearRegression(),x,y,cv=5,scoring='r2').mean())
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=96)
rfr.fit(x_train,y_train)
pre=rfr.predict(x_test)
print('r2 score is',r2_score(y_test,pre))
print('RMSE is ',np.sqrt(mean_squared_error(y_test,pre)))
print('Cross Validation Score is ',cross_val_score(RandomForestRegressor(),x,y,cv=5,scoring='r2').mean())
```

```python
from sklearn.ensemble import AdaBoostRegressor
abr=AdaBoostRegressor()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=96)
abr.fit(x_train,y_train)
pre=abr.predict(x_test)
print('r2 score is',r2_score(y_test,pre))
print('RMSE is ',np.sqrt(mean_squared_error(y_test,pre)))
print('Cross Validation Score is ',cross_val_score(abr,x,y,cv=5,scoring='r2').mean())
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=96)
gbr.fit(x_train,y_train)
pre=gbr.predict(x_test)
print('r2 score is',r2_score(y_test,pre))
print('RMSE is ',np.sqrt(mean_squared_error(y_test,pre)))
print('Cross Validation Score is ',cross_val_score(abr,x,y,cv=5,scoring='r2').mean())
```
